<h2>Generate Summary Statistics<h2>

```python
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")
display(df.describe())
```

<h2>Create Histograms and Boxplots for Numeric Features<h2>

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
```

```python
df[numeric_cols].hist(bins=20, figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numeric Features')
plt.tight_layout()
plt.show()

# Boxplots
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=col, data=df)
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

<h2>Pairplot and Correlation Matrix<h2>

```python
plt.figure(figsize=(10, 6))
corr = df[numeric_cols + ['Survived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

sns.pairplot(df[numeric_cols + ['Survived']], hue='Survived', palette='Set1')
plt.suptitle("Pairplot of Numeric Features by Survival", y=1.02)
plt.show()
```

<h2>Identify Patterns, Trends, or Anomalies<h2>

```python
print(
    "\nPatterns, Trends, or Anomalies:\n"
    "- Outliers are visible in Fare and Age from boxplots.\n"
    "- Fare distribution is right-skewed: many passengers paid low fares.\n"
    "- Survivors tend to have paid higher fares and were often younger.\n"
    "- Higher SibSp and Parch counts show possible family travel patterns."
)
```

<h2>Make Basic Feature-Level Inferences from Visuals<h2>

```python
print(
    "\nFeature-Level Inferences:\n"
    "- Higher Fare correlates with higher survival (possibly 1st class).\n"
    "- Age shows a wider range among non-survivors.\n"
    "- Passengers with fewer siblings/spouses (SibSp) and parents/children (Parch) had mixed survival chances.\n"
    "- Class and gender are likely major survival predictors (can be confirmed with more plots)."
)
```
